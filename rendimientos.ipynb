# Rendimientos


```python
import pandas as pd
import yfinance as yf
import numpy as np

# definimos dataframe

df = yf.download(
    'TSLA' , # ticker
    start = '2010-01-01' , # fecha de entrada
    end = '2024-06-17' , # fecha de salida
    progress= False )  # progreso

df = df.loc[:,['Adj Close']] # con esto obtengo solo la columna adj close
df = df.loc["2020-05-05":,['Adj Close']] # con esto obtengo todos los adj close a partir de dicha fecha
df.rename(columns = {'Adj Close' : 'adj_close'} , inplace = True) # cambiamos el nombre de la columna adj close para evitar confusiones con los espacios
# Calculamos los rendimientos simples y logaritmicos

df['rendimiento_simple'] = df.adj_close.pct_change() # calculamos el rendimiento simple con la funcion pct_change()


df['rendimiento_logaritmico'] = np.log(df.adj_close/df.adj_close.shift(1)) # calculamos el rendimiento logartimico con el logaritmo de la division de los adj close

df
```
